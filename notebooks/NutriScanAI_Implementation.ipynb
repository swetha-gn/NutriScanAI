```python
# STEP 4 — DATA CLEANING AND PREPROCESSING (FINAL CLEAN VERSION)

import pandas as pd
import numpy as np
import re

def clean_text(text):
    """Basic text cleaning for ingredients or labels."""
    if pd.isna(text):
        return ""
    text = re.sub(r"[^a-zA-Z0-9,.\s]", " ", str(text))
    text = re.sub(r"\s+", " ", text).strip().lower()
    return text

def clean_dataframe(df, name):
    """Common cleaning operations across datasets."""
    print(f"Cleaning {name} ...")

    before = len(df)
    df = df.drop_duplicates().reset_index(drop=True)
    after = len(df)
    print(f"   → Removed {before - after} duplicates")

    null_cols = df.columns[df.isna().mean() > 0.95]
    if len(null_cols) > 0:
        print(f"   → Dropping mostly empty columns: {list(null_cols)[:5]}")
        df = df.drop(columns=null_cols)

    for col in df.columns:
        if df[col].dtype in [np.float64, np.int64]:
            df[col] = df[col].fillna(df[col].median())
        else:
            df[col] = df[col].fillna(df[col].mode()[0] if not df[col].mode().empty else "")

    return df


# Open Food Facts
df_openfood_clean = clean_dataframe(df_off, "Open Food Facts")
df_openfood_clean["ingredients_text"] = df_openfood_clean["ingredients_text"].apply(clean_text)
df_openfood_clean["labels"] = df_openfood_clean["labels"].apply(clean_text)

# Datafiniti / Rishi Damarla Ingredients
df_ingredients_clean = clean_dataframe(df_datafiniti, "Datafiniti Ingredients")
df_ingredients_clean["features.value"] = df_ingredients_clean["features.value"].apply(clean_text)

# Food Allergens (handle variable naming variations)
df_allergens_clean = clean_dataframe(df_allergens, "Food Allergens")

if "Allergens" in df_allergens_clean.columns:
    df_allergens_clean["Allergens"] = df_allergens_clean["Allergens"].apply(clean_text)
elif "allergens" in df_allergens_clean.columns:
    df_allergens_clean["allergens"] = df_allergens_clean["allergens"].apply(clean_text)

# Nutritional Values
df_nutri_values_clean = clean_dataframe(df_nutri_values, "Nutritional Values")
df_nutri_values_clean.columns = [c.strip().replace(" ", "_").lower() for c in df_nutri_values_clean.columns]

# Shruti Saxena Nutrition
df_food_nutrition_clean = clean_dataframe(df_food_nutrition, "Shruti Saxena Nutrition")
df_food_nutrition_clean.columns = [c.strip().replace(" ", "_").lower() for c in df_food_nutrition_clean.columns]

# Merge ingredient-related datasets
df_ingredients_merged = pd.concat([
    df_openfood_clean[["product_name", "ingredients_text"]].rename(columns={"product_name": "name"}),
    df_ingredients_clean[["name", "features.value"]].rename(columns={"features.value": "ingredients_text"})
], ignore_index=True).drop_duplicates(subset=["ingredients_text"]).reset_index(drop=True)

print("All datasets cleaned and unified successfully!")
print("Merged ingredient dataset shape:", df_ingredients_merged.shape)
```

```python
import re
import pandas as pd

def auto_label_ingredients_optimized(df, text_col="ingredients_text"):
    """
    Robust vegetarian / non-vegetarian / uncertain classification
    from noisy ingredient lists.

    Key features:
    - Uses semantic normalization (e.g., 'pork fat' == 'animal fat')
    - Handles negations ('no egg', 'egg-free')
    - Avoids false positives ('eggplant', 'beefsteak tomato')
    - Distinguishes additives (e.g., 'E441' → gelatin)
    - Supports multilingual keywords ('pollo', 'carne', 'gelatina', etc.)
    - Assigns 'uncertain' for unclear or mixed compositions
    """

    # Canonical lists (can be expanded or externalized)
    veg_keywords = {
        "milk", "cheese", "paneer", "butter", "curd", "yogurt", "soy", "tofu",
        "vegetable", "plant", "bean", "lentil", "corn", "rice", "fruit", "nut",
        "flour", "sugar", "herb", "spice", "grain", "seed", "oat", "wheat", "barley"
    }

    nonveg_keywords = {
        "egg", "gelatin", "fish", "chicken", "beef", "pork", "bacon", "shrimp",
        "crab", "meat", "lard", "animal fat", "sausage", "ham", "shellfish",
        "anchovy", "tuna", "goat", "mutton", "veal", "duck", "turkey", "clam",
        "octopus", "squid"
    }

    # Common E-number additives (European codes) mapped to non-veg origin
    animal_e_codes = {"e441": "gelatin", "e542": "edible bone phosphate",
                      "e901": "beeswax", "e904": "shellac", "e913": "lanolin"}

    # Spanish/Italian translations (common in imported items)
    nonveg_translations = {"carne", "pollo", "cerdo", "pescado", "gelatina", "jamon", "mariscos", "huevo"}

    # Compile regex for non-veg terms with boundaries
    nonveg_pattern = re.compile(r'\b(' + '|'.join(nonveg_keywords | nonveg_translations) + r')\b', flags=re.IGNORECASE)
    veg_pattern = re.compile(r'\b(' + '|'.join(veg_keywords) + r')\b', flags=re.IGNORECASE)

    # Negation patterns
    negation_phrases = re.compile(r'\b(no|without|free from|egg-free|meatless|vegan|vegetarian)\b', flags=re.IGNORECASE)

    def classify_ingredients(text):
        if pd.isna(text) or not str(text).strip():
            return "uncertain"

        text = text.lower()

        # Normalize special characters
        text = re.sub(r"[^a-z0-9\s]", " ", text)
        text = re.sub(r"\s+", " ", text).strip()

        # E-number detection
        e_matches = [code for code in animal_e_codes if code in text]

        # Detect negation zones (e.g., "no egg", "without meat")
        if negation_phrases.search(text):
            # Check if negation precedes a nonveg term
            if nonveg_pattern.search(text):
                return "vegetarian"

        # Keyword presence
        has_nonveg = bool(nonveg_pattern.search(text) or e_matches)
        has_veg = bool(veg_pattern.search(text))

        # Handle false positives like "eggplant", "beefsteak tomato"
        if re.search(r'\beggplant\b', text) or re.search(r'\bbeefsteak tomato\b', text):
            has_nonveg = False

        if has_nonveg and not has_veg:
            return "non-vegetarian"
        elif has_veg and not has_nonveg:
            return "vegetarian"
        elif has_veg and has_nonveg:
            return "uncertain"
        else:
            return "uncertain"

    df = df.copy()
    df["label"] = df[text_col].apply(classify_ingredients)
    print(df["label"].value_counts())

    return df
```

```python
def refine_uncertain_labels(df, text_col="ingredients_text", label_col="label"):
    """
    Reclassify previously 'uncertain' items based on contextual rules:
    - Neutral terms (salt, oil, spice) → vegetarian
    - Animal-related terms (meat, fish, seafood, snails, insects) → non-vegetarian
    """

    neutral_terms = {
        "salt", "water", "oil", "olive", "sunflower", "spice", "herb", "vinegar",
        "pepper", "sugar", "flavor", "extract", "seasoning", "acid", "yeast"
    }

    seafood_terms = {
        "fish", "crab", "shrimp", "squid", "snail", "clam", "oyster", "prawn", "octopus", "lobster"
    }

    meat_terms = {
        "beef", "chicken", "pork", "goat", "mutton", "veal", "lamb", "duck", "turkey", "ham"
    }

    animal_terms = seafood_terms | meat_terms | {"gelatin", "animal fat", "insect", "larvae", "snails"}

    def reclassify(text, current_label):
        text = str(text).lower().strip()

        # Skip if already confident
        if current_label != "uncertain":
            return current_label

        # If it has animal/seafood words
        if any(word in text for word in animal_terms):
            return "non-vegetarian"

        # If it’s purely neutral (contains only salt/oil/spice)
        tokens = set(re.findall(r"[a-z]+", text))
        if tokens and tokens.issubset(neutral_terms):
            return "vegetarian"

        # If contains words like 'organic', 'plant', 'grain', 'seed'
        if any(word in text for word in ["organic", "grain", "seed", "plant", "vegetable", "fruit"]):
            return "vegetarian"

        return current_label

    df[label_col] = df.apply(lambda x: reclassify(x[text_col], x[label_col]), axis=1)

    print("Updated label distribution:")
    print(df[label_col].value_counts())

    return df


# Apply refinement after optimized labeling
df_ingredients_refined = refine_uncertain_labels(df_ingredients_labeled)
```

```python
import re
import pandas as pd
from difflib import get_close_matches

# Define extended keyword sets
veg_keywords = {
    "milk", "cheese", "butter", "cream", "soy", "tofu", "vegetable", "bean",
    "lentil", "corn", "rice", "wheat", "flour", "fruit", "nut", "seed",
    "plant", "herb", "spice", "sugar", "oil", "oats", "barley", "grain",
    "coconut", "almond", "peanut", "cashew", "sesame", "flax", "sunflower",
    "potato", "tomato", "carrot", "beet", "broccoli", "lettuce", "mushroom"
}

nonveg_keywords = {
    "egg", "gelatin", "fish", "chicken", "beef", "pork", "bacon", "shrimp",
    "crab", "squid", "snail", "clam", "oyster", "prawn", "octopus", "lard",
    "anchovy", "meat", "veal", "mutton", "duck", "turkey", "animal fat",
    "beef extract", "whey protein", "casein", "rennet", "collagen"
}

neutral_words = {
    "salt", "water", "flavor", "vinegar", "acid", "vitamin", "additive",
    "extract", "enzyme", "preservative", "color", "glycerin", "starch"
}


def classify_with_confidence_v2(text):
    """Advanced reliable classifier with fuzzy, conflict, and confidence logic."""
    if pd.isna(text) or not str(text).strip():
        return "uncertain", 0.0

    text = text.lower()
    text = re.sub(r"[^a-z0-9\s]", " ", text)
    tokens = set(text.split())

    veg_hits = sum(1 for w in veg_keywords if w in tokens)
    nonveg_hits = sum(1 for w in nonveg_keywords if w in tokens)
    neutral_hits = sum(1 for w in neutral_words if w in tokens)

    # Handle fuzzy matches for plural/typo variations
    for word in tokens:
        if get_close_matches(word, veg_keywords, n=1, cutoff=0.9):
            veg_hits += 0.5
        if get_close_matches(word, nonveg_keywords, n=1, cutoff=0.9):
            nonveg_hits += 0.5

    total_hits = veg_hits + nonveg_hits

    # 1️⃣ If no clear info, mark uncertain
    if total_hits == 0:
        return "uncertain", 0.0

    # 2️⃣ Compute confidence
    confidence = round(abs(veg_hits - nonveg_hits) / max(total_hits, 1), 2)

    # 3️⃣ Label assignment logic
    if nonveg_hits > veg_hits:
        label = "non-vegetarian"
    elif veg_hits > nonveg_hits:
        label = "vegetarian"
    else:
        label = "uncertain"

    # 4️⃣ Contradiction / Conflict check
    if veg_hits > 0 and nonveg_hits > 0:
        label = "uncertain"
        confidence = min(confidence, 0.4)

    # 5️⃣ Auto-upgrade purely neutral → vegetarian
    if label == "uncertain" and tokens.issubset(neutral_words):
        label, confidence = "vegetarian", 0.6

    return label, confidence


# Apply to dataset
df_ingredients_refined["label"], df_ingredients_refined["confidence"] = zip(
    *df_ingredients_refined["ingredients_text"].apply(classify_with_confidence_v2)
)

# Review updated results
print(df_ingredients_refined["label"].value_counts())
print(df_ingredients_refined[["ingredients_text", "label", "confidence"]].sample(10, random_state=42))
```

```python
def classify_with_confidence_v3(text):
    """Advanced classifier with fuzzy matching, contradiction handling, and confidence scoring."""
    if pd.isna(text) or not str(text).strip():
        return "uncertain", 0.0

    text = text.lower()
    text = re.sub(r"[^a-z0-9\s]", " ", text)
    tokens = set(text.split())

    # Direct matches
    veg_hits = sum(1 for w in veg_keywords if w in tokens)
    nonveg_hits = sum(1 for w in nonveg_keywords if w in tokens)
    neutral_hits = sum(1 for w in neutral_words if w in tokens)

    # Fuzzy corrections for typos/plurals
    for word in tokens:
        if get_close_matches(word, veg_keywords, n=1, cutoff=0.9):
            veg_hits += 0.5
        if get_close_matches(word, nonveg_keywords, n=1, cutoff=0.9):
            nonveg_hits += 0.5

    total_hits = veg_hits + nonveg_hits

    # Base case: no clear signal
    if total_hits == 0:
        return "uncertain", 0.0

    # Compute confidence
    confidence = round(abs(veg_hits - nonveg_hits) / max(total_hits, 1), 2)

    # Initial label decision
    if nonveg_hits > veg_hits:
        label = "non-vegetarian"
    elif veg_hits > nonveg_hits:
        label = "vegetarian"
    else:
        label = "uncertain"

    # Conflict override (veg + nonveg both present)
    if veg_hits > 0 and nonveg_hits > 0:
        label = "uncertain"
        confidence = min(confidence, 0.4)

    # All-neutral correction → likely veg
    if label == "uncertain" and tokens.issubset(neutral_words):
        label, confidence = "vegetarian", 0.6

    return label, confidence
```

```python
def fuse_labels(row, conf_thresh=0.75):
    """
    Combines additive-based status with ingredient-text classification.
    Conservative logic to minimize false negatives for vegetarians.
    """
    veg_status = str(row.get("veg_status", "unknown")).lower()
    text_label = str(row.get("final_label", "uncertain")).lower()
    conf = row.get("prediction_confidence", 0.0)

    # Direct rules (hard overrides)
    if veg_status == "non-vegetarian":
        return "non-vegetarian"
    if text_label == "non-vegetarian" and conf >= 0.7:
        return "non-vegetarian"

    # Strong vegetarian evidence
    if veg_status == "vegetarian" and conf >= 0.5:
        return "vegetarian"
    if text_label == "vegetarian" and conf >= conf_thresh:
        return "vegetarian"

    # Weak or mixed signals → fallback
    if "gelatin" in str(row.get("ingredients_text", "")).lower():
        return "non-vegetarian"
    if any(x in str(row.get("ingredients_text", "")).lower() for x in ["milk", "butter", "cheese", "cream"]):
        return "vegetarian"

    return "uncertain"

# Apply fusion
df_openfood_clean["smart_label"] = df_openfood_clean.apply(fuse_labels, axis=1)

# Re-check distribution
print(df_openfood_clean["smart_label"].value_counts())
```

```python
import re
import numpy as np

def classify_product(text, additive_df=df_add_full):
    """
    Classifies an ingredient text into vegetarian / non-vegetarian / uncertain
    using both additive codes and keyword logic.
    """
    if pd.isna(text) or str(text).strip() == "":
        return {"label": "uncertain", "confidence": 0.0, "reason": "Empty text"}

    text_lower = text.lower()

    # 1. Detect additives (E-codes)
    additives = re.findall(r"e\d{3,4}[a-z]?", text_lower)
    additives = list(set(additives))

    # 2. Look up vegetarian/vegan status
    add_status = additive_df[additive_df["clean_code"].isin(additives)]
    veg_values = list(add_status["vegetarian"].dropna().str.lower().unique())

    # 3. Keyword detection
    veg_keywords = [
        "soy", "tofu", "bean", "rice", "wheat", "corn", "nut",
        "milk", "cheese", "butter", "yogurt", "fruit", "vegetable", "plant"
    ]
    nonveg_keywords = [
        "gelatin", "fish", "meat", "beef", "pork", "chicken",
        "shrimp", "crab", "anchovy", "egg", "lard", "tallow", "rennet"
    ]

    has_veg = any(word in text_lower for word in veg_keywords)
    has_nonveg = any(word in text_lower for word in nonveg_keywords)

    # 4. Rule-based fusion
    label, confidence, reason = "uncertain", 0.5, "No strong evidence"

    # Non-veg additive override
    if "no" in veg_values or has_nonveg:
        label, confidence = "non-vegetarian", 1.0
        reason = f"Detected non-veg keywords/additives: {', '.join(additives or [''])}"

    elif "yes" in veg_values or has_veg:
        label, confidence = "vegetarian", 0.9
        reason = f"Detected vegetarian indicators: {', '.join(additives or [''])}"

    elif "maybe" in veg_values:
        label, confidence = "uncertain", 0.6
        reason = f"Contains additives with uncertain origin: {', '.join(additives or [''])}"

    return {"label": label, "confidence": confidence, "reason": reason}
```

```python

# 0. SETUP

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    classification_report, confusion_matrix, ConfusionMatrixDisplay, accuracy_score
)
import re


df_final = df_openfood_clean[['ingredients_text', 'smart_label']].dropna()


# 1. DATA PREPARATION

X = df_final['ingredients_text'].astype(str)
y = df_final['smart_label']

print("Original class balance:")
print(y.value_counts())

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

print("\nTrain/Test sizes:", len(X_train), "/", len(X_test))


# 2. TF-IDF VECTORIZATION

vectorizer = TfidfVectorizer(
    max_features=8000,
    ngram_range=(1,2),
    stop_words='english'
)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

print("TF-IDF matrix:", X_train_vec.shape)


# 3. MULTICLASS CLASSIFIER (3-class: veg / non-veg / uncertain)

clf = LogisticRegression(
    max_iter=3000,
    class_weight='balanced',
    multi_class='ovr',
    solver='liblinear'
)
clf.fit(X_train_vec, y_train)


# 4. EVALUATION

y_pred = clf.predict(X_test_vec)

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred, labels=clf.classes_)
ConfusionMatrixDisplay(cm, display_labels=clf.classes_).plot(cmap="YlGnBu")
plt.title("Smart Label Classification (TF-IDF + Logistic Regression)")
plt.show()

print("Accuracy:", round(accuracy_score(y_test, y_pred),3))


# 5. FEATURE IMPORTANCE

feature_names = np.array(vectorizer.get_feature_names_out())
coefs = clf.coef_

for idx, cls in enumerate(clf.classes_):
    top_pos = np.argsort(coefs[idx])[-15:]
    top_neg = np.argsort(coefs[idx])[:15]
    print(f"\n🔹 Class: {cls}")
    print("Top Positive Indicators:", feature_names[top_pos])
    print("Top Negative Indicators:", feature_names[top_neg])


# 6. PREDICTION FUNCTION

def ml_predict_ingredient(text):
    vec = vectorizer.transform([text])
    label = clf.predict(vec)[0]
    prob = round(clf.predict_proba(vec).max(), 3)
    return {"text": text, "predicted_label": label, "confidence": prob}

examples = [
    "gelatin, sugar, E441",
    "soy milk, sugar, cocoa butter",
    "vitamin D2, sunflower oil",
    "E120, glucose syrup, natural color"
]

preds = [ml_predict_ingredient(e) for e in examples]
pd.DataFrame(preds)
```

```python

# 0. SETUP

import pandas as pd
import numpy as np
import re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

df_final = df_openfood_clean[['ingredients_text', 'smart_label']].dropna()


# 1. SEPARATE RULE-CONFIDENT & UNCERTAIN ROWS

# df_uncertain_to_classify: items that are *still* uncertain after initial rules
df_uncertain_to_classify = df_final[df_final["smart_label"] == "uncertain"].copy()

# df_ml_train_data: items that were initially classified as vegetarian or non-vegetarian
# This data will be used to train the ML model to resolve df_uncertain_to_classify
df_ml_train_data = df_final[df_final["smart_label"] != "uncertain"].copy()

print(f"Total samples: {len(df_final)}")
print(f"ML training data (confident): {len(df_ml_train_data)}")
print(f"To be classified by ML (uncertain): {len(df_uncertain_to_classify)}")


# 2. TRAIN ML MODEL ON CONFIDENT DATA (TO RESOLVE UNCERTAIN)

# Ensure there's enough data and at least two classes in the training data
if len(df_ml_train_data) > 20 and len(df_ml_train_data['smart_label'].unique()) > 1:
    X_train_ml, X_test_ml, y_train_ml, y_test_ml = train_test_split(
        df_ml_train_data["ingredients_text"], df_ml_train_data["smart_label"],
        test_size=0.2, random_state=42, stratify=df_ml_train_data["smart_label"]
    )

    vectorizer_ml = TfidfVectorizer(
        max_features=5000,
        ngram_range=(1,2),
        stop_words="english"
    )
    X_train_vec_ml = vectorizer_ml.fit_transform(X_train_ml)
    X_test_vec_ml = vectorizer_ml.transform(X_test_ml)

    clf_resolver = LogisticRegression(
        max_iter=2000,
        class_weight="balanced",
        solver="liblinear",
        multi_class='ovr' # Handles multiple classes (veg/non-veg)
    )
    clf_resolver.fit(X_train_vec_ml, y_train_ml)

    y_pred_ml = clf_resolver.predict(X_test_vec_ml)
    print("\n📊 Evaluation of ML resolver on confident subset:")
    print(classification_report(y_test_ml, y_pred_ml))
    print("Accuracy:", round(accuracy_score(y_test_ml, y_pred_ml), 3))

else:
    print("⚠️ Not enough confident samples with multiple classes to train ML model reliably.")
    clf_resolver = None
    vectorizer_ml = None


# 3. RULE-BASED LOGIC (STRONG SIGNALS)

nonveg_terms = [
    "gelatin", "e441", "e542", "e631", "e627", "meat", "fish", "beef",
    "pork", "egg", "shrimp", "lard", "tallow", "isinglass", "rennet", "cochineal", "e120"
]
veg_terms = [
    "soy", "tofu", "bean", "vegetable", "plant", "wheat", "flour",
    "milk", "butter", "cheese", "nut", "fruit", "rice", "corn", "lentil", "vegan", "vegetarian"
]

def rule_based_label(text):
    t = str(text).lower() # Ensure text is string
    if any(re.search(rf"\b{term}\b", t) for term in nonveg_terms):
        return "non-vegetarian", 1.0, "Rule match: non-veg keyword"
    if any(re.search(rf"\b{term}\b", t) for term in veg_terms):
        return "vegetarian", 1.0, "Rule match: veg keyword"
    return "uncertain", 0.5, "No rule match"


# 4. HYBRID PREDICT FUNCTION

def hybrid_predict(text):
    rule_label, rule_conf, rule_reason = rule_based_label(text)
    if rule_label != "uncertain":
        return {
            "text": text,
            "final_label": rule_label,
            "confidence": rule_conf,
            "source": "rule",
            "reason": rule_reason
        }
    elif clf_resolver is not None and vectorizer_ml is not None:
        # Use the resolver model trained on confident data to classify uncertain items
        vec = vectorizer_ml.transform([text])
        ml_label = clf_resolver.predict(vec)[0]
        ml_conf = clf_resolver.predict_proba(vec).max()
        return {
            "text": text,
            "final_label": ml_label,
            "confidence": round(ml_conf, 3),
            "source": "ml_resolver",
            "reason": "ML resolver inferred from confident data"
        }
    else:
        return {
            "text": text,
            "final_label": "uncertain",
            "confidence": 0.5,
            "source": "none",
            "reason": "No ML model available or trained"
        }


# 5. TEST HYBRID LOGIC

samples = [
    "sugar, milk solids, cocoa butter, soy lecithin, vanilla extract",
    "gelatin, sugar, flavoring, E441",
    "E120, glucose syrup, natural color", # E120 is Cochineal, should be non-veg
    "vitamin D2, sunflower oil",
    "refined sunflower oil, vitamin A, vitamin D2",
    "egg white powder, sugar, vanilla essence"
]

results = [hybrid_predict(s) for s in samples]
pd.DataFrame(results)
```

```python

# 0. INSTALL DEPENDENCIES

!pip install -q transformers datasets torch accelerate evaluate


# 1. IMPORTS AND SETUP

import pandas as pd
import torch
from datasets import Dataset, DatasetDict
from sklearn.model_selection import train_test_split
from transformers import (
    DistilBertTokenizerFast,
    DistilBertForSequenceClassification,
    TrainingArguments,
    Trainer
)
import evaluate
import numpy as np


# 2. LOAD AND PREPARE DATA

df = df_openfood_clean[['ingredients_text', 'smart_label']].dropna().reset_index(drop=True)
df = df[df['smart_label'].isin(['vegetarian','non-vegetarian','uncertain'])]

label2id = {'vegetarian':0, 'non-vegetarian':1, 'uncertain':2}
id2label = {v:k for k,v in label2id.items()}
df['label'] = df['smart_label'].map(label2id)

train_df, test_df = train_test_split(df, test_size=0.2, stratify=df['label'], random_state=42)
train_ds = Dataset.from_pandas(train_df[['ingredients_text','label']])
test_ds  = Dataset.from_pandas(test_df[['ingredients_text','label']])
dataset = DatasetDict({'train':train_ds, 'test':test_ds})


# 3. TOKENIZE

model_name = "distilbert-base-uncased"
tokenizer = DistilBertTokenizerFast.from_pretrained(model_name)

def tokenize_fn(batch):
    return tokenizer(batch["ingredients_text"], truncation=True, padding="max_length", max_length=128)

tokenized_ds = dataset.map(tokenize_fn, batched=True)
tokenized_ds = tokenized_ds.remove_columns(["ingredients_text"])


# 4. LOAD MODEL

model = DistilBertForSequenceClassification.from_pretrained(
    model_name,
    num_labels=3,
    id2label=id2label,
    label2id=label2id
)


# 5. TRAINING SETUP

metric = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    acc = metric.compute(predictions=preds, references=labels)
    return {"accuracy": acc["accuracy"]}

training_args = TrainingArguments(
    output_dir="./bert_food_classifier",
    eval_strategy="epoch", # Corrected argument name
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
    load_best_model_at_end=True,
    logging_dir="./logs",
    report_to="none"
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_ds["train"],
    eval_dataset=tokenized_ds["test"],
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)


# 6. TRAIN

trainer.train()


# 7. EVALUATE

metrics = trainer.evaluate()
print("📊 Evaluation metrics:", metrics)


# 8. SAMPLE PREDICTIONS

texts = [
    "gelatin, sugar, E441",
    "soy milk, sugar, cocoa butter",
    "vitamin D2, sunflower oil",
    "E120, glucose syrup, natural color",
    "refined sunflower oil, vitamin A, vitamin D2"
]

inputs = tokenizer(texts, return_tensors="pt", padding=True, truncation=True, max_length=128)
with torch.no_grad():
    outputs = model(**inputs)
probs = torch.nn.functional.softmax(outputs.logits, dim=-1)
preds = torch.argmax(probs, dim=-1)

for t, p, pr in zip(texts, preds, probs):
    print(f"\n🧾 {t}\n→ Label: {id2label[p.item()]} (Confidence: {pr.max().item():.2f})")
```
